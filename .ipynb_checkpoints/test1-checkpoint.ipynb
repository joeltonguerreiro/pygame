```python
import pygame
import sys

SCREEN_SIZE = WIDTH, HEIGHT = (320, 240)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 50, 50)
GREEN = (50, 255, 50)
BLUE = (0, 0, 255)

# Initialization
pygame.init()
screen = pygame.display.set_mode(SCREEN_SIZE)
pygame.display.set_caption('Circles')
fps = pygame.time.Clock()
paused = False


square = pygame.Rect(300, 230, 20, 20)

INITIAL_VELOCITY = 0.1

velocity_x = INITIAL_VELOCITY
velocity_y = INITIAL_VELOCITY


left_pad = pygame.Rect(10, 210, 20, 60)
right_pad = pygame.Rect(600, 210, 20, 60)

pads = [left_pad, right_pad]


def update(velocity_x, velocity_y):
    square.move_ip(velocity_x, velocity_y)
    
    for pad in pads:
        if square.colliderect(pad):
            if velocity_x > 0:
                square.right = pad.left
                
            if velocity_x < 0:
                square.left = pad.right
                
            if velocity_y > 0:
                square.bottom = pad.top
                
            if velocity_y < 0:
                square.top = pad.bottom
    
    
def render():
    screen.fill(BLACK)
        
    for pad in pads:
        pygame.draw.rect(screen, WHITE, pad)
    
    pygame.draw.rect(screen, RED, square)
        
    pygame.display.flip()
    

while True:
    delta_time = fps.tick(60)
    
    for event in pygame.event.get():
        velocity_x = 0
        velocity_y = 0
    
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
            
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_SPACE:
                paused = not paused
                
    keys = pygame.key.get_pressed()
    
    if keys[pygame.K_UP]:
        velocity_y = -(INITIAL_VELOCITY * delta_time)
        #print('y', velocity_y, delta_time)

    if keys[pygame.K_DOWN]:
        velocity_y = (INITIAL_VELOCITY * delta_time)
        #print('y', velocity_y, delta_time)

    if keys[pygame.K_LEFT]:
        velocity_x = -(INITIAL_VELOCITY * delta_time)
        #print('x', velocity_x, delta_time)

    if keys[pygame.K_RIGHT]:
        velocity_x = (INITIAL_VELOCITY * delta_time)
        #print('x', velocity_x, delta_time)
        
    
    if not paused:
        update(velocity_x, velocity_y)
        render()
```
