```python
import pygame
import sys

SCREEN_SIZE = WIDTH, HEIGHT = (320, 240)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 50, 50)
GREEN = (50, 255, 50)
BLUE = (0, 0, 255)

# Initialization
pygame.init()
screen = pygame.display.set_mode(SCREEN_SIZE)
pygame.display.set_caption('Circles')
fps = pygame.time.Clock()
paused = False


square = pygame.Rect(30, 50, 20, 20)

INITIAL_VELOCITY = 0.1

velocity_x = INITIAL_VELOCITY
velocity_y = INITIAL_VELOCITY

walls = []

blocks = [
    "WWWWWWWWWWWWWWWW",
    "W              W",
    "W              W",
    "W              W",
    "W              W",
    "W              W",
    "W              W",
    "W              W",
    "W              W",
    "W              W",
    "W              W",
    "WWWWWWWWWWWWWWWW",
]

left = 0
top = 0

for row in blocks:
    for col in row:
        if col == "W":
            wall = pygame.Rect(left, top, 20, 20)
            walls.append(wall)
        left += 20
    top += 20
    if left >= 320:
        left = 0


def update(velocity_x, velocity_y):
    square.move_ip(velocity_x, velocity_y)
    
    for wall in walls:
        if square.colliderect(wall):
            if velocity_x > 0:
                square.right = wall.left
                
            if velocity_x < 0:
                square.left = wall.right
                
            if velocity_y > 0:
                square.bottom = wall.top
                
            if velocity_y < 0:
                square.top = wall.bottom
    
    
def render():
    screen.fill(BLACK)
        
    for wall in walls:
        pygame.draw.rect(screen, WHITE, wall)
    
    pygame.draw.rect(screen, RED, square)
        
    pygame.display.flip()
    

while True:
    delta_time = fps.tick(60)
    
    for event in pygame.event.get():
        velocity_x = 0
        velocity_y = 0
    
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
            
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_SPACE:
                paused = not paused
    
    #captura as teclas pressionadas
    keys = pygame.key.get_pressed()
    
    if keys[pygame.K_UP]:
        velocity_y = -(INITIAL_VELOCITY * delta_time)

    if keys[pygame.K_DOWN]:
        velocity_y = (INITIAL_VELOCITY * delta_time)

    if keys[pygame.K_LEFT]:
        velocity_x = -(INITIAL_VELOCITY * delta_time)

    if keys[pygame.K_RIGHT]:
        velocity_x = (INITIAL_VELOCITY * delta_time)
        
    
    if not paused:
        update(velocity_x, velocity_y)
        render()
```
